```python
df = pd.read_csv('../data/fusion_def_clean.csv', dtype=df_dtypes, low_memory=False)
df.head(10)
```

```python
############################### Préparation des données - Suppression des colonnes inutiles ###########################
```

```python
columns_to_drop = ['Unnamed: 0', 'senc', 'locp', 'actp', 'etatp', 'an_nais', 'an_naiss', 'age_acc', 'age_acc_seconds', 'age_acc_an', "permis"]
df1 = df.drop(columns=columns_to_drop, axis=1)
df1.head(10)
```

```python
df1.info()
```

```python
# Transformation des variables catégorielles (qualitatives) en variables "quantitatives" afin qu'elles puissent être
# interprétées par un modèle de machine learning.

# 1.Traitement des variables catégorielles nominales : One Hot Encoding (dichotomisation) sur les variables suivantes : 
# 'agg', 'int', 'atm', 'col', 'catr', 'circ', 'vosp', 'prof', 'plan', 'surf', 'infra', 'situ', 'obs', 'obsm',
# 'choc', 'manv', 'catv_Label', 'catu', 'sexe', 'trajet', 'secuUn', 'secuDeux'
```

```python
vosp = df1['vosp']
df1 = df1.join(pd.get_dummies(vosp, prefix='vosp'))
```

```python
agg = df1['agg']
df1 = df1.join(pd.get_dummies(agg, prefix='agg'))
```

```python
int = df1['int']
df1 = df1.join(pd.get_dummies(int, prefix='int'))
```

```python
atm = df1['atm'] 
df1 = df1.join(pd.get_dummies(atm, prefix='atm'))
```

```python
col = df1['col'] 
df1 = df1.join(pd.get_dummies(col, prefix='col'))
```

```python
catr = df1['catr'] 
df1 = df1.join(pd.get_dummies(catr, prefix='catr'))
```

```python
circ = df1['circ'] 
prof = df1['prof'] 
plan = df1['plan'] 

df1 = df1.join(pd.get_dummies(circ, prefix='circ'))
df1 = df1.join(pd.get_dummies(prof, prefix='prof'))
df1 = df1.join(pd.get_dummies(plan, prefix='plan'))
```

```python
surf = df1['surf'] 
infra = df1['infra'] 
situ = df1['situ'] 

df1 = df1.join(pd.get_dummies(surf, prefix='surf'))
df1 = df1.join(pd.get_dummies(infra, prefix='infra'))
df1 = df1.join(pd.get_dummies(situ, prefix='situ'))
```

```python
obs = df1['obs'] 
obsm = df1['obsm'] 
choc = df1['choc']

df1 = df1.join(pd.get_dummies(obs, prefix='obs'))
df1 = df1.join(pd.get_dummies(obsm, prefix='obsm'))
df1 = df1.join(pd.get_dummies(choc, prefix='choc'))
```

```python
catu = df1['catu'] 
sexe = df1['sexe'] 
trajet = df1['trajet'] 

df1 = df1.join(pd.get_dummies(catu, prefix='catu'))
df1 = df1.join(pd.get_dummies(sexe, prefix='sexe'))
df1 = df1.join(pd.get_dummies(trajet, prefix='trajet'))
```

```python
secuUn = df1['secuUn'] 
secuDeux = df1['secuDeux'] 

df1 = df1.join(pd.get_dummies(secuUn, prefix='secuUn'))
df1 = df1.join(pd.get_dummies(secuDeux, prefix='secuDeux'))
```

```python
catv_Label = df1['catv_Label'] 
df1 = df1.join(pd.get_dummies(catv_Label, prefix='catv'))
```

```python
manv = df1['manv'] 
df1 = df1.join(pd.get_dummies(manv, prefix='manv'))
```

```python
df1.info()
```

```python
# Conservation des variables indicatrices correspondants aux différentes modalités des variables qualitatives dichotomisées
# et suppression des variables qualitatives concernées.
df1 = df1.drop(['agg', 'int', 'atm', 'col', 'catr', 'circ', 'vosp', 'prof', 'plan', 'surf', 'infra', 'situ', 'obs', 'obsm', 
                'choc', 'manv', 'catv_Label', 'catu', 'sexe', 'trajet', 'secuUn', 'secuDeux'], axis=1)
```

```python
# Suppression de la variable date qui n'apporte aucune information supplémentaire par rapport aux variables 'jour', 'mois'
# et 'annee_x_x'

df1 = df1.drop('date', axis=1)
```

```python
df1.head()
```

```python
# 2.Conversion des variables catégorielles ordinales et de la variable cible en variables numériques
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df1['lum'] = le.fit_transform(df['lum'])
df1['num_veh'] = le.fit_transform(df['num_veh'])
df1['place'] = le.fit_transform(df['place'])
df1['tranches_ages'] = le.fit_transform(df['tranches_ages'])
df1['grav'] = le.fit_transform(df['grav'])
```

```python
df1.info()
```

```python
for col in df1.columns:
    print(col, df1[col].dtype)    
```

```python
# Conversion des variables 'com' et 'dep' (object/string) en float 
```

```python
df1['com'] = df1['com'].str.replace(r'(2A)', '200', regex=True).replace(r'(2B)', '201', regex=True).astype('float')
```

```python
df1['com'].dtype
```

```python
df1['dep'] = df1['dep'].str.replace(r'(2A)', '200', regex=True).replace(r'(2B)', '201', regex=True).astype('float')
```

```python
df1['dep'].dtype
```

```python
# Affectation des "features" du dataset dans data et affectation de la variable cible dans "target"  
```

```python
data = df1.drop('grav', axis=1)
```

```python
target = df1['grav']
```

```python
# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
################################# Construction du modèle de classification ######################################
```

```python
# Création du classifieur et entrainement du modèle sur les données d'entraînement
clf = linear_model.LogisticRegression(C=1.0) 
clf.fit(X_train, y_train)
```

```python
################################ Évaluation du modèle de classification #########################################
```

```python
# Calcul des prédictions pour les données présentes dans l'ensemble de test
y_pred = clf.predict(X_test)
```

```python
# Calcul de la matrice de confusion - Méthode 1 : à l'aide de sklearn
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Calcul de la matrice de confusion - Méthode 2 : à l'aide de pandas
cm = pd.crosstab(y_test, y_pred, rownames=['Classe réelle'], colnames=['Classe prédite'])
cm
```

```python
################################################# Commentaires #####################################################

# la matrice de confusion nous montre que la totalité des usagers accidentés de l'ensemble de test ont été classés
# comme appartenant à la classe 2 (blessé hospitalisé) alors que cette classe ne concerne en réalité que 41 % 
# du niveau de gravité des accidents.
# Par conséquent, notre modèle est aussi performant qu'un modèle purement aléatoire qui attribuerait une classe
# en la choisissant au hasard uniformément. Il apparait donc nécessaire de régler les hyperparamètres spécifiques
# au modèle de régression logistique afin d'obtenir le meilleur modèle possible.
# Nous utiliserons la fonction GridSearchCV puis nous effectuerons une réévaluation de la performance moyenne
# du modèle associé par validation croisée.

# Cependant, autant que je sache, la régression logistique ne présente pas d'hyperparamètres véritablement critiques. 
# Par conséquent, lors d'une prochaine étape, je suggère plutôt d'appliquer des pénalités voire recourir à un solver
# autre que de type Newton.

# Remarque générale : le modèle de classification par régression logistique est un classificateur binaire
# pour déterminer la probabilité qu'une observation appartienne à une classe spécifique. 
# Aussi, ce n'est sans doute pas le meilleur classificateur pour notre problématique de ML. 
```

```python
# Calcul du taux de bonnes prédictions du modèle.
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
# Affichage du rapport de classification pour évaluer les performances par classe
print("Rapport de classification :\n", classification_report(y_test, y_pred))
```

```python
################################################# Commentaires #####################################################

# le rapport de classification précise :
# Que la précision et le rappel sont nuls pour les classes 0, 1 et 3. 
# Ce qui signifie que ces classes n'ont pas du tout été bien gérées par le modèle.

# Concernant la classe 2 (blessé hospitalisé), la précision est inférieure à la moyenne (41 %), 
# toutefois le rappel est élevé (100 %).
# Ce qui signifie que la classe a été bien gérée par le modèle. Celle-ci a bien été détectée, 
# cependant elle inclut également des observations d'autres classes. 

# Concernant le F1-score, celui-ci est de 58 %. Il a en commun avec l’accuracy de résumer la performance d’un modèle
# en un seul indicateur pour chaque seuil de classification. Il est plus complexe mais a l’avantage d’être
# robuste en présence de données déséquilibrées... 

# En conclusion, le modèle n'est pas fiable.

# Par ailleurs, nous pouvons effectivement constater que notre dataset est déséquilibré.
# Nous avons un déséquilibre flagrant avec la classe 3 (tué) qui représente que 2.66 % des observations. 
# ==> Nous allons appliquer 2 méthodes de rééchantillonnage.

# 1. Nous appliquerons tout d'abord la méthode Oversampling qui permet d'augmenter le nombre d'observations des classes
# minoritaires afin d'arriver à un ratio classes minoritaires / classe majoritaire satisfaisant.

# 2. Puis nous appliquerons la méthode Undersampling qui permet de diminuer le nombre d'observations des classes
# majoritaires afin d'arriver à un ratio classes minoritaires / classe majoritaire satisfaisant.
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler, ClusterCentroids
from imblearn.metrics import classification_report_imbalanced, geometric_mean_score
from sklearn.metrics import f1_score
from sklearn.svm import SVC
```

```python
# 1. Application de la méthode Random Oversampling
rOs = RandomOverSampler()
X_ro, y_ro = rOs.fit_resample(X_train, y_train)
print('Classes échantillon oversampled :', dict(pd.Series(y_ro).value_counts()))
```

```python
# Entraînement du modèle sur l'ensemble issu du RandomOversampler

clf.fit(X_ro, y_ro)
y_pred = clf.predict(X_test)
```

```python
# Affichage de la matrice de confusion obtenue grâce aux prédictions sur l'ensemble de test

pd.crosstab(y_test, y_pred, rownames=['Classe réelle'], colnames=['Classe prédite'])
```

```python
# Calcul du taux de bonnes prédictions du modèle.
clf.score(X_test, y_test)
```

```python
################################################# Commentaires #####################################################

# la matrice de confusion nous montre que la totalité des usagers accidentés de l'ensemble de test ont été classés
# comme appartenant à la classe 1 (blessé léger) alors que cette classe ne concerne en réalité que 35.81 % des accidents.
# Par conséquent, notre modèle est encore aussi performant qu'un modèle purement aléatoire qui attribuerait une classe
# en la choisissant au hasard uniformément.
```

```python
# Rapport de résultat sur l'ensemble de test obtenu par classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_pred))
```

```python
############################################## Commentaires ##################################################

# la précision et le rappel sont nuls pour les classes 0, 2 et 3. 
# Ces classes n'ont pas du tout été bien gérées par le modèle.

# Concernant la classe 1, la précision est inférieure à la moyenne (36 %), toutefois le rappel est élevé (100 %).
# Cette fois-ci, la classe 1 a bien été gérée par le modèle, celle-ci a bien été détectée.
# Cependant elle inclut également des observations d'autres classes. 

# En conclusion, notre modèle n'est toujours pas fiable. La méthode de rééchantillonage Oversampling ne permet pas
# d'améliorer les performances de notre modèle. La majorité des classes n'étant pas bien gérées.
```

```python
# 2. Application de la méthode Random Undersampling
```

```python
# Random Undersampling
rUs = RandomUnderSampler()
X_ru, y_ru = rUs.fit_resample(X_train, y_train)
print('Classes échantillon undersampled :', dict(pd.Series(y_ru).value_counts()))
```

```python
# Entraînement du modèle sur l'ensemble issu du RandomUnderSampler 

clf.fit(X_ru, y_ru)
y_pred = clf.predict(X_test)
```

```python
# Affichage de la matrice de confusion obtenue grâce aux prédictions sur l'ensemble de test

print(pd.crosstab(y_test, y_pred, rownames=['Classe réelle'], colnames=['Classe prédite']))
```

```python
# Calcul du taux de bonnes prédictions du modèle.
clf.score(X_test, y_test)
```

```python
############################################## Commentaires ##################################################

# la matrice de confusion nous montre que la totalité des usagers accidentés de l'ensemble de test ont été classés
# comme appartenant à la classe 2 (blessé hospitalisé) alors que cette classe ne concerne en réalité que 35.81 % des accidents.
# Par conséquent, notre modèle est encore aussi performant qu'un modèle purement aléatoire qui attribuerait une classe
# en la choisissant au hasard uniformément.

# Le taux de bonnes prédictions du modèle (41%) est rigoureusement identique à celui obtenu sans rééchantillonage
# des observations.
```
